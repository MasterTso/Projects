```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

plt.style.use('default')
```

```python
data_path = "Downloads/Finance_data.csv"
df = pd.read_csv(data_path)

df.head()
```

```python
df.info()
```

```python
df.shape, df.columns
```

```python
df.describe()
```

```python
numerical_cols = df.select_dtypes(include=np.number)

mean_values = numerical_cols.mean()
median_values = numerical_cols.median()
mode_values = numerical_cols.mode().iloc[0]

mean_values, median_values, mode_values
```

```python
variance = numerical_cols.var()
std_dev = numerical_cols.std()
data_range = numerical_cols.max() - numerical_cols.min()

variance, std_dev, data_range
```

```python
numerical_cols.hist(figsize=(15, 10), bins=30)
plt.suptitle("Distribution of Financial Variables", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=numerical_cols)
plt.xticks(rotation=45)
plt.title("Box Plot of Financial Variables")
plt.show()
```

```python
sample_col = numerical_cols.columns[0]
sample_data = numerical_cols[sample_col]

probability = np.mean(sample_data > sample_data.mean())
probability
```

```python
z_scores = stats.zscore(sample_data)
z_scores[:10]
```

```python
corr_matrix = numerical_cols.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
t_stat, p_value = stats.ttest_1samp(sample_data, sample_data.mean())

t_stat, p_value
```
